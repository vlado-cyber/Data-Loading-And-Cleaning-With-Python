## Removing redundant/unnecessary columns

```python
df.drop(['BusinessTravel','DailyRate','EmployeeCount','HourlyRate',
         'JobInvolvement','NumCompaniesWorked','Over18','OverTime',
         'RelationshipSatisfaction','StockOptionLevel','TotalWorkingYears',
         'YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager'],
        axis=1, inplace=True)
```

## Rename columns

```python
df.rename({"MonthlyIncome":"EmployeeSalary", "MonthlyRate":"EmployeeRevenue"}, axis=1, inplace=True)
```

## Add new columns

```python
df["Age<=25"] = df["Age"] <= 25
df["Age<=25"].value_counts()
```

```python
df["Age>=55"] = df["Age"] >=55
```

```python
df["ROI"] = (df["EmployeeRevenue"] - df["EmployeeSalary"] / df["EmployeeSalary"]) * 100
```

```python
df
```

## Replace values in columns

```python
education_map = {1:"High School", 2:"Undergraduate", 3:"Postgraduate", 4:"Professional Certification", 5:"Other"}
env_satisfaction_map = {1:"Dissatisfied", 2:"Neutral", 3:"Satisfied", 4:"Very Satisfied"}
job_level_map = {1:"Entry-Level", 2:"Junior", 3:"Mid-level", 4:"Senior", 5:"Executive"}
job_satisfaction_map = {1:"Dissatisfied", 2:"Neutral", 3:"Satisfied", 4:"Very Satisfied"}
standard_hours_map = {80:40}
```

```python
df["Education"] = df["Education"].map(education_map)
df["EnvironmentSatisfaction"] = df["EnvironmentSatisfaction"].map(env_satisfaction_map)
df["JobLevel"] = df["JobLevel"].map(job_level_map)
df["JobSatisfaction"] = df["JobSatisfaction"].map(job_satisfaction_map)
df["StandardHours"] = df["StandardHours"].map(standard_hours_map)
```

```python
df.to_csv('cleaned_hr_data.csv', index=False)
```
